Van der Slycken & Bleys. (Economic Welfare Measures)

```python
#BCE(Benefits and Costs Experienced)
#Importing data & Marking Acronym

BCE_Data = pd.read_excel(r'/WISE Database/Data/Source Data/VSB-EWM/IX-W-BCE-JvdS.xlsx', sheet_name='W-BCE')

BCE_Data['Indicator'] = 'VSB-EWM-BCE'

BCE_Data.rename(columns={'Value BCE':'Value'},inplace= True)

BCE_Data['Country']=BCE_Data['Country'].astype(str)

BCE_Data
```
